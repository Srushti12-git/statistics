Question 1: What is the difference between descriptive statistics and inferential
statistics? Explain with examples.
Ans:-
Descrptive Statistics:
Descriptive statistics simply describe, summarize, and show the features of a dataset. They allow you to present the data in a meaningful way so that patterns can emerge. They do not allow for conclusions beyond the data analyzed.
Example: A Class of Students
Imagine you have the final exam scores for all 30 students in a specific math class.
Calculation: You calculate the average (mean) score to be 85%.
Measure: You find the standard deviation is 5%, indicating most scores are clustered closely around the mean.
Representation: You create a histogram to show the distribution of the scores.
Conclusion: You can confidently state that, for this specific class, the average score was 85%. You cannot, however, use this to say anything about a different class or all students in the school without using inferential methods.

Inferential Statistics:
Inferential statistics use data from a sample to draw conclusions, make predictions, or test hypotheses about a larger population from which the sample was drawn. This process always involves a degree of uncertainty, so results are often expressed in terms of probability.

Example: All High School Students

Imagine a state department of education wants to know the average SAT score for all 11th graders in the state (the population). It's impractical to test every student, so they select a random sample of 500 students.

Hypothesis Testing: They use the average score of the 500-student sample to test the hypothesis that the state-wide average is greater than a certain historical value.

Estimation: They calculate a 95% confidence interval that estimates the true average SAT score for all 11th graders in the state.

Conclusion: Based on the sample, they might infer, "We are 95% confident that the true average SAT score for all 11th graders in the state is between 1050 and 1150." The conclusion is a generalization about the entire population.


Question 2: What is sampling in statistics? Explain the differences between random
and stratified sampling.
Ans:-
The two core probability sampling methods, Simple Random Sampling (SRS) and Stratified Sampling, differ primarily in their execution and when they are best applied to a population.

1.Simple Random Sampling (SRS)
i.SRS is defined by its Process: Members are selected randomly from the entire population, often using a random number generator. The Goal of this method is straightforward: to give every single member of the population an equal chance of being selected, thus ensuring the sample is generally representative of the overall group and minimizing bias.

ii.Suitability: This technique is Best for homogeneous populations where all members share similar characteristics, meaning there are no significant subgroups whose representation must be specifically ensured.

iii.Advantage: Its main strength lies in its Simplicity to implement and its inherent low risk of researcher bias.

iv.Example: To sample a university population of 10,000 students, you simply put all 10,000 names into a hat and randomly draw 500 names.

2.Stratified Sampling
i.Stratified Sampling is a more structured approach. Its Process involves first dividing the population into non-overlapping subgroups called strata based on relevant characteristics, and then drawing a random sample from each stratum. The Goal is to guarantee representation from these specific, key subgroups within the population, which is crucial when the population is diverse (heterogeneous).

ii.Suitability: This method is Best for heterogeneous populations where distinct subgroups are expected to have different responses or characteristics relevant to the study.

iii.Advantage: It typically provides more precise estimates and is superior for making comparisons between different subgroups.

iv.Example: To ensure all class levels are represented when sampling 10,000 students, you would first divide the population into four strata (Freshmen, Sophomores, Juniors, and Seniors). You would then randomly select a proportional number of students from each of those four groups to form the final sample.

Question 3: Define mean, median, and mode. Explain why these measures of central tendency are important.
Ans:-
üìê Measures of Central Tendency: Mean, Median, and Mode
Measures of Central Tendency are single values that attempt to describe a set of data by identifying the central position within that set. They provide a concise summary of the distribution's typical or average value.

i.Mean
->The value obtained by summing all data points and then dividing by the total number of data points.
->$\bar{x} = \frac{\sum x}{n}$ (where $\sum x$ is the sum of all values and $n$ is the number of values).

ii.Median
->The middle value in a data set that has been ordered from smallest to largest.
->Find the $(n+1)/2$ position. If $n$ is odd, it's the middle value. If $n$ is even, it's the average of the two middle values.

iii.Mode
->The value that occurs most frequently in a data set.
->Count the frequency of each value and identify the one with the highest count. A data set can have one mode, more than one mode (bimodal/multimodal), or no mode.

Question 4: Explain skewness and kurtosis. What does a positive skew imply about the data?
Ans:-
Skewness: Measuring Asymmetry
Skewness is a measure of the asymmetry of a probability distribution. It indicates the extent to which the data points are shifted or clustered toward one side, resulting in a longer tail on the other side.
->Zero Skewness: The distribution is perfectly symmetrical (like a normal or "bell-shaped" curve), and the mean, median, and mode are all equal.

->Positive Skew (Right Skew): The tail extends to the right (the positive direction).

->Negative Skew (Left Skew): The tail extends to the left (the negative direction).

Kurtosis: Measuring Tail Heaviness
Kurtosis is a measure of the "tailedness" of a distribution relative to a normal distribution. It describes how much of the data is concentrated in the center (peak) and in the tails, which impacts the likelihood of outliers (extreme values).

->Mesokurtic: The distribution has kurtosis similar to a normal distribution.
->Leptokurtic: The distribution has a sharper peak and heavier tails. This means there is a higher probability of extreme events or outliers.
->Platykurtic: The distribution has a flatter peak and lighter tails. This indicates the data is more spread out, and extreme values are less likely.


A positive skew (or right-skewed distribution) has specific implications for the data's shape and its central tendency measures:
1.Tail Direction: The distribution has a long tail extending toward the right (positive direction) of the graph.
2.Data Concentration: The bulk of the data is clustered on the left side (lower values) of the distribution.
3.Outliers: The presence of a few extremely large values (outliers) in the right tail pulls the average upward.
4.Relationship of Averages: The mean, median, and mode are ordered as follows:
Mode< Median < Mean
a.The Mode (most frequent value) is the smallest because the most common values are clustered on the left.
b.The Median (middle value) is next.
c.The Mean (average) is the largest because it is most affected by the few high-value outliers on the right.

Real-World Example: The distribution of individual incomes in a country is typically positively skewed. Most people earn a smaller to moderate salary (clustering on the left), but a few extremely high earners (the long right tail) significantly inflate the average (mean) income.

```python
#Question 5: Implement a Python program to compute the mean, median, and mode of a given list of numbers.
#numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]

import numpy as np
import statistics
numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]

#calculating the mean
mean_of_numbers=np.mean(numbers)
print(f"The mean of numbers is :{mean_of_numbers}")

#calculating the median
median_of_numbers=np.median(numbers)
print(f"The median of numbers is :{median_of_numbers}")

#calculating the mode
mode_of_numbers=statistics.mode(numbers)
print(f"The mode of numbers is :{mode_of_numbers}")
```

```python
#Question 6: Compute the covariance and correlation coefficient between the following two datasets provided as lists in Python:
#list_x = [10, 20, 30, 40, 50]
#list_y = [15, 25, 35, 45, 60]

import numpy as np
list_x=[10,20,30,40,50]
list_y=[15,25,35,45,60]

covariance_of_list=np.cov(list_x,list_y)
print(f"The covariance of 2 lists is {covariance_of_list}")

correlation_of_list=np.corrcoef(list_x,list_y)
print(f"The correlation coefficient of 2 lists is {correlation_of_list}")
```

```python
#Question 9: Your team has collected customer satisfaction survey data on a scale of 1-10 and wants to understand its distribution before launching a new product.
#‚óè Explain which summary statistics and visualizations (e.g. mean, standard deviation, histogram) you‚Äôd use.
#‚óè Write Python code to create a histogram using Matplotlib for the survey data:
#survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore') #suppress all the warnings

survey_scores=[7,8,5,9,7,8,9,10,4,7,6,9,8,7]
survey_series=pd.Series(survey_scores)
mean_score=survey_series.mean()
median_score=survey_series.median()
std_dev=survey_series.std()
skewness=survey_series.skew()
mode_score=survey_series.mode()

print("--- Summary Statistics ---")
print(f"Total Responses (n): {len(survey_scores)}")
print(f"Mean Score: {mean_score:.2f}")
print(f"Median Score: {median_score:.2f}")
print(f"Mode Score: {mode_score.tolist()}")
print(f"Standard Deviation: {std_dev:.2f}")
print(f"Skewness: {skewness:.2f}")

plt.figure(figsize=(10,6))
plt.hist(survey_scores,bins=np.arange(3.5, 11.5))
plt.title('Customer Satisfaction survey data')
plt.xlabel('Satisfaction Score (Data Value)')
plt.ylabel('Frequency (Count)')
plt.grid(axis='y', alpha=0.5, linestyle='--')


```
