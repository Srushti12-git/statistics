Question 1: What is hypothesis testing in statistics?

Ans:-Hypothesis testing is a statistical method used as a form of inference to decide whether there is enough evidence in a sample of data to reject a specific, tentative assumption (or claim) made about a population parameter or a population probability distribution. It's a formal procedure for investigating ideas about the world using statistics.

It allows researchers and analysts to use sample data to make inferences and draw conclusions about a larger population, quantifying the likelihood that observed differences or relationships occurred by chance.

Question 2: What is the null hypothesis, and how does it differ from the alternative hypothesis?

Ans:
What is Null hypothesis?

The null hypothesis ($H_0$) is the statement of no effect, no difference, or no relationship.It serves as the default assumption or the status quo that is assumed to be true until there is sufficient statistical evidence to suggest otherwise.

->Status Quo: It represents the position that no change has occurred or that any observed difference is simply due to random chance or sampling error.

->Mathematical Form: It always includes an equality component (typically $=$, but can be $\ge$ or $\le$).Example: The average height ($\mu$) of adult men is equal to 70 inches. $H_0: \mu = 70$

->Testing Goal: The goal of a statistical test is to determine whether the sample data provides enough evidence to reject the null hypothesis.
Analogy: Think of the legal system's principle of "innocent until proven guilty." The null hypothesis (15$H_0$) is the assumption of "innocence" (no effect), and the burden of proof is on the alternative hypothesis.

How Does it Differ from the Alternative Hypothesis ($H_a$ or $H_1$)?

The alternative hypothesis ($H_a$ or $H_1$) is the statement that the researcher is trying to find evidence to support.19 It directly contradicts the null hypothesis.
| Feature | Null Hypothesis ($H_0$) | Alternative Hypothesis ($H_a$ or $H_1$) |
| :--- | :--- | :--- |
| **Definition** | A statement of **no effect**, no difference, or no relationship in the population. The status quo. | A statement that **there is an effect**, a difference, or a relationship in the population. The research claim. |
| **Symbol** | $H_0$ (H-naught or H-zero) | $H_a$ or $H_1$ |
| **Key Symbol** | Always includes **equality** ($=$, $\le$, or $\ge$) | Always includes **inequality** ($\ne$, $<$, or $>$ ) |
| **Goal with Data** | You attempt to **reject** it. | You attempt to **support** it. |
| **Conclusion** | You either **Reject** $H_0$ or **Fail to Reject** $H_0$. | $H_a$ is **supported** when $H_0$ is rejected. |

Question 3: Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.

Ans:The significance level, denoted by the Greek letter $\alpha$ (alpha), is a crucial threshold set by the researcher before conducting the hypothesis test.It fundamentally defines the acceptable risk tolerance for making a certain type of error.

Defining the Significance Level

The significance level $\alpha$ is formally defined as the probability of rejecting the null hypothesis ($H_0$) when, in reality, $H_0$ is true:
$$\alpha = P(\text{Reject } H_0 \mid H_0 \text{ is true})$$

->Type I Error (False Positive): This is the error of concluding that a significant effect or difference exists when it does not (i.e., you reject $H_0$ when you shouldn't have).

->Common Values: The most frequently used significance level is $\alpha = 0.05$ (5%).Other common values are $\alpha = 0.01$ (1%) or $\alpha = 0.10$ (10%).

->Risk Tolerance: Choosing a smaller 15$\alpha$ (e.g., 0.01) makes the test more stringent, reducing the risk of a Type I Error, but it increases the risk of a Type II Error (failing to reject 16$H_0$ when it is false).

Role in Deciding the Outcome

The significance level plays a direct and non-negotiable role in determining the outcome of a statistical test by setting the threshold for rejection. This is achieved through two equivalent methods:

1. The P-Value Approach (Most Common)

The p-value is calculated from the sample data and represents the probability of observing results as extreme as, or more extreme than, the current data, assuming the null hypothesis ($H_0$) is true.

The decision rule is:

| Feature | Null Hypothesis ($\mathbf{H_0}$) | Alternative Hypothesis ($\mathbf{H_a}$ or $\mathbf{H_1}$) |
| :--- | :--- | :--- |
| **P-value $\le \alpha$** | **Reject the Null Hypothesis ($\mathbf{H_0}$)** | The result is statistically significant because the observed data is considered too unlikely to have occurred by random chance if $H_0$ were true. |
| **P-value $ > \alpha$** | **Fail to Reject the Null Hypothesis ($\mathbf{H_0}$)** | The result is not statistically significant because the observed data is reasonably likely to have occurred by random chance if $H_0$ were true. |

2. The Critical Value Approach (Equivalently Defined by $\alpha$)

The significance level $\alpha$ defines the critical region (or rejection region) of the test statistic's distribution.

->The critical region is the area under the probability curve where the test statistic must fall for $H_0$ to be rejected.The total area of this region equals $\alpha$.

->A critical value is the boundary of this rejection region.

->Decision Rule: If the calculated test statistic (e.g., Z-score, t-score) falls into the critical region (i.e., past the critical value), $H_0$ is rejected.If it does not, we fail to reject $H_0$.

Question 4: What are Type I and Type II errors? Give examples of each.

Ans:-In hypothesis testing, since we are using sample data to make a decision about a population, there is always a chance of making an incorrect conclusion. These incorrect conclusions are categorized into two types of errors: Type I and Type II.

1. Type I Error (False Positive)

A Type I Error occurs when the researcher rejects a null hypothesis ($H_0$) that is actually true in the population.

->Error Description: Concluding there is a significant effect, difference, or relationship when there isn't one in reality. It is a "false alarm."

->Alternative Name: False Positive

->Probability: The probability of a Type I Error is denoted by the Greek letter 7$\alpha$ (alpha), which is the pre-set significance level of the test.
$P(\text{Type I Error}) = \alpha$

Example of Type I Error: Medical Test

| Feature | Null Hypothesis (H₀) | Alternative Hypothesis (Hₐ or H₁) |
| :---: | :---: | :--- |
| Hypothesis Set-up | H₀: Person does not have the disease. | Decision: Reject H₀ (Test is positive) |
| Truth & Error | Reality: H₀ is True (Person does not have the disease) | Error: Type I Error (False Positive) |
| Consequence | The patient is incorrectly told they have a serious disease, potentially leading to unnecessary anxiety, costly follow-up tests, or harmful treatments. | *N/A* |

2. Type II Error (False Negative)
A Type II Error occurs when the researcher fails to reject a null hypothesis ($H_0$) that is actually false in the population.

->Error Description: Failing to detect a significant effect, difference, or relationship that does exist in reality. It is a "missed opportunity."

->Alternative Name: False Negative

->Probability: The probability of a Type II Error is denoted by the Greek letter $\beta$ (beta).

=>$P(\text{Type II Error}) = \beta$
=>The Power of a test is 15$1 - \beta$, which is the probability of correctly rejecting a false $H_0$.

Example of Type II Error: Drug Trial

| Feature | Null Hypothesis ($\mathbf{H_0}$) | Alternative Hypothesis ($\mathbf{H_a}$ or $\mathbf{H_1}$) |
| :---: | :---: | :--- |
| **Hypothesis Set-up** | $H_0$: Drug is **not** effective. | $H_a$: The new drug is **effective**. |
| **Decision & Reality** | **Decision:** Fail to Reject $H_0$ (Conclude drug is not effective) | **Reality:** $H_a$ is True (The drug is effective) |
| **Error & Consequence** | **Error:** Type II Error (False Negative) | A genuinely effective new drug is mistakenly abandoned, causing patients to miss out on a beneficial treatment. |

Question 5: What is the difference between a Z-test and a T-test? Explain when to use each.

Ans:-
| Feature | **Z-Test** | **T-Test** |
| :---: | :---: | :---: |
| **Primary Condition** | Population Standard Deviation ($\sigma$) is Known | Population Standard Deviation ($\sigma$) is Unknown |
| **Sample Size (n)** | Best for Large Samples ($n \ge 30$) | Best for Small Samples ($n < 30$) |
| **Distribution Used** | Standard Normal Distribution (Z-distribution) | Student's T-Distribution |
| **Standard Error** | Uses $\sigma$ (Population SD) | Uses $s$ (Sample SD, an estimate) |
| **Formula (One-Sample)** | $Z = \frac{\bar{x} - \mu}{\sigma / \sqrt{n}}$ | $t = \frac{\bar{x} - \mu}{s / \sqrt{n}}$ |
| **Degrees of Freedom** | Not applicable (Z-distribution is fixed) | Required ($df = n-1$ for one sample) |

1. Use a Z-Test:
A Z-test is the appropriate choice when you have sufficient information to assume that the sample mean follows a normal distribution with a known population variance.

->Scenario A: Known Population Standard Deviation ($\sigma$)

If you know the true standard deviation of the entire population ($\sigma$), you should use the Z-test, regardless of the sample size. (This is rare in real-world research but common in textbook problems or well-established quality control.)

->Scenario B: Large Sample Size ($n \ge 30$)

If the population standard deviation ($\sigma$) is unknown, but the sample size is large (typically $n \ge 30$), the Central Limit Theorem suggests the sampling distribution of the mean is approximately normal.7 In this case, the sample standard deviation (8$s$) is a reliable substitute for $\sigma$, and the test statistic approximates the Z-distribution.For simplicity, many statisticians still use the Z-test in this large-sample, unknown $\sigma$ case.

2. Use a T-Test:
A T-test is designed to handle the additional variability and uncertainty that comes from estimating the population parameter from a limited amount of data.

->Primary Scenario: Unknown Population Standard Deviation ($\sigma$) AND Small Sample Size ($n < 30$)

If the true population standard deviation ($\sigma$) is unknown, and you are working with a small sample, the T-test must be used.The T-distribution corrects for the increased uncertainty by adjusting its shape based on the degrees of freedom ($n-1$).

->Default Choice: Since the population standard deviation is rarely known in practical research, the T-test is the most commonly used test for comparing means.

Question 6: Write a Python program to generate a binomial distribution with n=10 and p=0.5, then plot its histogram.

```python
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

#Define parameters
n = 10      # Number of trials
p = 0.5     # Probability of success
N = 10000   # Number of experiments (sample size for simulation)

# Generate the binomial distribution sample
# The result is the number of successes in n trials, repeated N times
binomial_sample = np.random.binomial(n, p, N)

# Plot the histogram
plt.figure(figsize=(10,6))
bins=np.arange(n+2)-0.5
plt.hist(binomial_sample,bins=bins,density=True,color='skyblue',edgecolor='black',alpha=0.75)

#Set labels, title, and grid
plt.title(f'Histogram of Binomial Distribution (n={n},p={p})')
plt.xlabel('Number of Successes (k)')
plt.ylabel('Probability/Relative Frequency')
plt.xticks(np.arange(n + 1)) # Ensure ticks are at integer values 0, 1, ..., 10
plt.grid(axis='y', alpha=0.5)
plt.show()
```

Question 8: Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, t
from math import sqrt

def calculate_ci_stats(data, confidence=0.95, known_sigma=None):
    """
    Calculates the confidence interval statistics for the population mean.

    Uses Z-test logic if known_sigma is provided, otherwise uses T-test logic
    (which is the robust default when sigma is unknown).
    
    Returns:
        tuple: (lower_bound, upper_bound, sample_mean, margin_of_error, test_type)
    """
    n=len(data)
    sample_mean=np.mean(data)
    if known_sigma is not None:
        # Case 1: Z-Test (Population sigma is KNOWN or n is large)
        std_dev = known_sigma
        dist = norm
        # Z-score for the given confidence level
        critical_value=dist.ppf(1-(1-confidence)/2)
        df = None
        test_type = "Z-Test (Sigma Known)"
        
    else:
        # Case 2: T-Test (Population sigma is UNKNOWN or n is large)
        std_dev=np.std(data,ddof=1)
        df=n-1
        dist=t
        #t-score for given confidence level and degree of freedom
        critical_value=dist.ppf(1-(1-confidence)/2,df=df)
        test_type = f"T-Test (Sigma Unknown, df={df})"
    
    #find margin_error    
    standard_error=std_dev/sqrt(n)
    margin_error=critical_value*standard_error
    
    #confidence interval
    lower_bound=sample_mean-margin_error
    upper_bound=sample_mean+margin_error
    
    return lower_bound,upper_bound,sample_mean,margin_error,test_type

def simulate_and_plot_confidence_interval(mu_pop, sigma_pop, sample_size, confidence=0.95, known_sigma=None):
    """
    Simulates data from a Normal Distribution, calculates the Confidence Interval, 
    and plots the results.
    """
    # --- 1. Data Simulation ---
    data = np.random.normal(loc=mu_pop, scale=sigma_pop, size=sample_size)
    
    # --- 2. Confidence Interval Calculation ---
    lower, upper, mean_samp, me, test_type = calculate_ci_stats(
        data, 
        confidence=confidence, 
        known_sigma=known_sigma
    )
    # --- 3. Print Results ---
    print("\n" + "=" * 60)
    print(f"--- Analysis for Sample of Size n={sample_size} ---")
    print(f"Population Parameters (Simulated): μ={mu_pop}, σ={sigma_pop}")
    print(f"Sample Mean (x̄): {mean_samp:.3f}")
    print(f"Test Type Used: {test_type}")
    print(f"Margin of Error (ME): ±{me:.3f}")
    print(f"{int(confidence*100)}% CI: ({lower:.3f}, {upper:.3f})")
    print("=" * 60)

    # --- 4. Plotting ---
    plt.figure(figsize=(10, 6))
    
    # Histogram
    plt.hist(
        data, 
        bins=int(np.sqrt(sample_size)), # Simple bin rule
        density=True, 
        color='#90EE90', 
        edgecolor='black', 
        alpha=0.7, 
        label='Sample Data Distribution'
    )
    
    # True Population Mean (Reference Line)
    plt.axvline(mu_pop, color='red', linestyle='-', linewidth=2, label=f'True Population Mean (μ={mu_pop})')
    
    # Sample Mean (Central Point)
    plt.axvline(mean_samp, color='blue', linestyle='--', linewidth=2, label=f'Sample Mean (x̄={mean_samp:.3f})')
    
    # Confidence Interval Bounds
    plt.axvline(lower, color='green', linestyle=':', linewidth=3, label=f'CI Lower Bound ({lower:.3f})')
    plt.axvline(upper, color='green', linestyle=':', linewidth=3, label=f'CI Upper Bound ({upper:.3f})')
    
    # Title and Labels
    plt.title(f'{int(confidence*100)}% Confidence Interval vs. Sample Data (Test: {test_type.split("(")[0]})')
    plt.xlabel('Value')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(axis='y', alpha=0.5)
    plt.show()

# --- DEMONSTRATION 1: Large Sample (n=100), Sigma Unknown (T-TEST is robust) ---
# We assume sigma is unknown, so the function uses the T-distribution logic.
simulate_and_plot_confidence_interval(
    mu_pop=65, 
    sigma_pop=10, 
    sample_size=100, 
    known_sigma=None # Treat sigma as unknown
)

# --- DEMONSTRATION 2: Small Sample (n=15), Sigma KNOWN (Z-TEST logic) ---
# This is rare, but demonstrates Z-logic if you explicitly provide sigma.
simulate_and_plot_confidence_interval(
    mu_pop=120, 
    sigma_pop=5, 
    sample_size=15, 
    known_sigma=5 # Provide the known population sigma
)
```
